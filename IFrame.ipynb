# Part 2: Working with Colors & Text

Here's how to create gradients with pixel manipulation:

````python
# Here's a colorful gradient example
width, height = 300, 300
img = Image.new('RGB', (width, height))

# This creates a simple gradient - try to understand how!
for x in range(width):
    for y in range(height):
        r = int((x / width) * 255)
        g = int((y / height) * 255)
        b = 100
        img.putpixel((x, y), (r, g, b))

plt.imshow(img)
plt.axis('off')
plt.show()
````

````python
from PIL import Image
import matplotlib.pyplot as plt
import noise  # pip install noise

width, height = 500, 250
scale = 0.01

img = Image.new('RGB', (width, height))

for x in range(width):
    for y in range(height):
        n = noise.pnoise2(x * scale, y * scale, octaves=6)
        n = (n + 1) / 2  # Normalize to 0-1

        # Boost blue channel more, keep reds moderate for purple tone
        r = int(130 + n * 80)    # 130 to 210 (purple red)
        g = int(n * 40)          # small green for richness
        b = int(220 + n * 35)    # 220 to 255 (strong blue)

        r = min(r, 255)
        g = min(g, 255)
        b = min(b, 255)

        img.putpixel((x, y), (r, g, b))

plt.imshow(img)
plt.axis('off')
plt.show()
````

## Challenge 6: Generative Pattern ðŸŽ¯

**Your task:** Create a pattern using loops and randomness. Try making:
* Random shapes with random colors
* A spiral pattern  
* Repeating geometric designs

````python
import random
import math
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

width, height = 500, 500
img = Image.new('RGB', (width, height), 'black')
draw = ImageDraw.Draw(img)

center_x, center_y = width // 2, height // 2
num_points = 150
angle_step = 0.3
radius_step = 2

for i in range(num_points):
    angle = i * angle_step
    radius = i * radius_step

    x = center_x + int(radius * math.cos(angle))
    y = center_y + int(radius * math.sin(angle))

    size = random.randint(10, 25)
    color = (random.randint(100, 255), random.randint(50, 200), random.randint(150, 255))

    bbox = [x - size//2, y - size//2, x + size//2, y + size//2]
    draw.ellipse(bbox, fill=color)

plt.imshow(img)
plt.axis('off')
plt.show()
````
